## Creating the First Model

```python
df1 = df.copy(deep=True)
```

```python
plt.figure(figsize=(20,10))
mask = np.triu(np.ones_like(df1.corr(numeric_only=True), dtype=bool))
sns.heatmap(df1.corr(numeric_only=True),mask=mask,annot=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df1['Attrition'] = le.fit_transform(df1['Attrition'])
```

```python
corr = df1.corr()
plt.figure(figsize=(20,5))
corr['Attrition'].sort_values(ascending=False)[1:].plot(kind="bar")
```

```python
df1.info()
```

```python
cols_to_dummy = []

for col in df1.columns:
    if (df1[col].dtype == 'object') & (col != 'Attrition'):
        cols_to_dummy.append(col)
```

```python
cols_to_dummy
```

```python
for c in cols_to_dummy:
    temp = pd.get_dummies(df1[c],drop_first=True)
    df1.drop(c,axis=1,inplace=True)
    df1 = pd.concat([df1,temp],axis=1)
```

```python
df1.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
X = df1.drop('Attrition',axis=1)
y = df1['Attrition']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
dt_cls = DecisionTreeClassifier()
rf_cls = RandomForestClassifier()
lg_cls = LogisticRegression()
```

```python
classifiers = [dt_cls,rf_cls,lg_cls]
```

```python
for c in classifiers:
    c.fit(x_train,y_train)
    pred1 = c.predict(x_train)
    pred2 = c.predict(x_test)
    cm = confusion_matrix(y_test,pred2,normalize='all')
    plt.figure(figsize=(10,5))
    sns.heatmap(cm,annot=True)
    plt.title(c)
    plt.xlabel("Actual")
    plt.ylabel("Predicted")
    print(classification_report(y_test,pred2))
    print("The training accuracy score for", c, "is", accuracy_score(y_train,pred1))
    print("The testing accuracy score for", c, "is", accuracy_score(y_test,pred2))
```
