## Determine Correlation and Columns to Drop

```python
df
```

```python
test1 = df.copy(deep=True)
```

```python
cols_to_delete = []
for i in test1.columns:
    if test1[i].dtype == "object":
        cols_to_delete.append(i)
        
cols_to_delete
```

```python
for i in cols_to_delete[1:]:
    test1.drop(i, axis=1)
```

```python
encoded_cols2 = pd.get_dummies(df['BusinessTravel'], drop_first=True)
encoded_cols3 = pd.get_dummies(df['Department'], drop_first=True)
encoded_cols4 = pd.get_dummies(df['EducationField'], drop_first=True)
encoded_cols5 = pd.get_dummies(df['Gender'], drop_first=True)
encoded_cols6 = pd.get_dummies(df['JobRole'], drop_first=True)
encoded_cols7 = pd.get_dummies(df['MaritalStatus'], drop_first=True)
encoded_cols8 = pd.get_dummies(df['OverTime'], drop_first=True)
```

```python
cols_to_add = [encoded_cols2,encoded_cols3,encoded_cols4,encoded_cols5,encoded_cols6,encoded_cols7,encoded_cols8]

for i in cols_to_add:
    merged1 = pd.concat([test1,i], axis=1)
```

```python
for i in cols_to_delete[1:]:
    merged1 = merged1.drop(i, axis=1)
```

```python
merged1
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
merged1['Attrition'] = le.fit_transform(merged1['Attrition']) 
```

```python
plt.figure(figsize=(20,10))
mask = np.triu(np.ones_like(merged1.corr(), dtype=bool))
sns.heatmap(merged1.corr(),mask=mask,annot=True)
```

drop columns:joblevel, performancerating, age, yearsatcompany, yearssincelastpromotion.

```python
merged1 = merged1.drop(['EmployeeID','JobLevel','PerformanceRating','Age','YearsAtCompany','YearsSinceLastPromotion'], axis=1)
```

```python
corr = merged1.corr()
plt.figure(figsize=(20,5))
corr['Attrition'].sort_values(ascending=False)[1:].plot(kind="bar");
```

## Build Model

```python
X = merged1.drop('Attrition', axis=1)
y = merged1['Attrition']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_class = RandomForestClassifier()
```

```python
rf_class.fit(x_train, y_train)
```

```python
rf_pred = rf_class.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,rf_pred)
```
